```python
filepath="/content/drive/My Drive/MachineLearningUdemy/Section2/Data2.csv"
df = pd.read_csv(filepath)
df
```

```python
X=df.drop(["Survived"], axis=1)
X=X.values
print(X)

Y= df["Survived"].values
print(Y)
```

```python
def is_numeric(element):
    try:
        float(element)
        return True
    except ValueError:
        return False

numeric_mask = np.vectorize(is_numeric)(X[0])
non_numeric_indices = np.where(~numeric_mask)[0]
non_numeric_indices
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct= ColumnTransformer(transformers =[("encoder", OneHotEncoder(), non_numeric_indices)], remainder="passthrough")
XCopy= np.array(ct.fit_transform(X))
XCopy
```
